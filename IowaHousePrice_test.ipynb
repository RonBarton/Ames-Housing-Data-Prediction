```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
pd.set_option('display.max_columns', 82)
```

# Imputation / Missing Values

```python
#fill in garage year built NAN with year built of house

df.loc[df['GarageYrBlt'].isnull(), 'GarageYrBlt'] = df.loc[df['GarageYrBlt'].isnull(), 'YearBuilt']
df['GarageYrBlt'] = df['GarageYrBlt'].astype('int')
```

```python
#keeping column labels for imputation...fancy impute needs an array and will lose labels

df_train_numeric_for_labels = df.select_dtypes(exclude=[object])
```

```python
import fancyimpute as fp
from fancyimpute import KNN

#impute continuous vars using KNN

df_train_numeric = df.select_dtypes(exclude=[object]).as_matrix()
df_complete1=pd.DataFrame(KNN().fit_transform(df_train_numeric), columns=df_train_numeric_for_labels.columns, index =df_train_numeric_for_labels.index)
```

```python
#check imputed fill
df_complete1.info()
```

```python
#Drop continuous vars from original dataframe then reaatach the dataframe that has the imputed values
df_drop_setB = df.select_dtypes(include=[object])

#df_complete1 is the dataframe w/ imputed vars
df_1 = df_complete1.join(df_drop_setB)
```

```python
#check data set after joining imputed values
df_1.info()
```

```python
#show only missing values
df_1.isna().any()[lambda x: x]
```

```python
#massvnry - masonry type there is already a none cateogry
#alley - NA is no alley
df_1.loc[df_1['Alley'].isnull(), 'Alley'] = 'NoAcc'
df_1.loc[df_1['MasVnrType'].isnull(), 'MasVnrType'] = 'None'

#NA means no basement
df_1[['BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2']] = \
df_1[['BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2']].fillna(value='NoBsmt')

#electrical
df_1.loc[df_1['Electrical'].isnull(), 'Electrical'] = 'SBrkr'
#NA means no firepalce
df_1.loc[df_1['FireplaceQu'].isnull(), 'FireplaceQu'] = 'NoFirepl'

#NA means no garage
df_1[['GarageType','GarageFinish','GarageQual','GarageCond']] = \
df_1[['GarageType','GarageFinish','GarageQual','GarageCond']].fillna(value='NoGarage')


#NA means no for poolqc, fence, miscfeature
df_1.loc[df_1['PoolQC'].isnull(),'PoolQC'] = 'NoPool'
df_1.loc[df_1['Fence'].isnull(),'Fence'] = 'NoFence'
df_1.loc[df_1['MiscFeature'].isnull(),'MiscFeature'] = 'None'
```

```python
#FINAL CHECK - show only missing values
df_1.isna().any()[lambda x: x]
```

# feature engineering

```python
# Converting all columns with categorical values into binary columns
#MSSubclass is a range of values that represent categories
df_1.loc[df_1['MSSubClass']==20,'MS_1S_New'] = 1
df_1.loc[df_1['MSSubClass']==30,'MS_1S_New'] = 0
df_1.loc[df_1['MSSubClass']==40,'MS_1S_Attic'] = 1
df_1.loc[df_1['MSSubClass']==45,'MS_1_5S_Fin'] = 0
df_1.loc[df_1['MSSubClass']==50,'MS_1_5S_Fin'] = 1
df_1.loc[df_1['MSSubClass']==60,'MS_2S_New'] = 1
df_1.loc[df_1['MSSubClass']==70,'MS_2S_New'] = 0
df_1.loc[df_1['MSSubClass']==75,'MS_2_5S'] = 1
df_1.loc[df_1['MSSubClass']==80,'MS_M_Lvl'] = 1
df_1.loc[df_1['MSSubClass']==85,'MS_SplitFoyer'] = 1
df_1.loc[df_1['MSSubClass']==90,'MS_Duplex'] = 1
df_1.loc[df_1['MSSubClass']==120,'MS_1S_PUD'] = 1
df_1.loc[df_1['MSSubClass']==160,'MS_2S_PUD'] = 1
df_1.loc[df_1['MSSubClass']==180,'MS_M_Lvl_PUD'] = 1
df_1.loc[df_1['MSSubClass']==190,'MS_2_Fam'] = 1
```

```python
#fill NAns with zero from created columns above
df_1.fillna(value=0, inplace=True)
df_1 = df_1.drop(columns='MSSubClass')
```

```python
df_1.info()
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler

# log transform Y
y_temp = df_1[['SalePrice']]
y_temp['SalePrice'] = np.log(y_temp['SalePrice'])

#drop y
X_1 = df_1.drop(columns='SalePrice')

#Drop categorical vars for scaling
X_1_drop = X_1.select_dtypes(exclude=[object])

# Standardise data
ss = StandardScaler()
X_cont = ss.fit_transform(X_1_drop)
X_cont = pd.DataFrame(X_cont, columns=X_1_drop.columns)
```

```python
#create df that includes the cateogrical vars to reattach
X_1_obj = X_1.select_dtypes(include=[object])

#join scaled continuous and cateogrical
df_2 = X_cont.join(X_1_obj)
```

```python
#join df_2 with log tranformed y
df_3= df_2.join(y_temp)
df_3.info()
```

# Encoding & Test/Train Split

```python
df_encode = pd.get_dummies(df_3)
```

```python
df_encode.info()
```

```python
from sklearn.model_selection import train_test_split
yy = df_encode[['SalePrice']]
XX = df_encode.drop(columns=['SalePrice','Id'])


#test/train split A
X_train2, X_test2, y_train2, y_test2 = train_test_split(XX, yy, test_size=0.2, random_state=42)

print(X_train2.shape)
print(y_train2.shape)
print(X_test2.shape)
print(y_test2.shape)
```

# LAsso

```python
#Check a lasso regression to see if it provides better results
#performs automatic feature selection
import numpy as np
from sklearn.linear_model import LassoCV
l1_cv = LassoCV(cv=5, max_iter=10000, alphas=np.linspace(0.0001, .1,num=1000))
lasso = l1_cv.fit(X_train2, y_train2)
```

```python
print('alpha = ', l1_cv.alpha_) 
print('coef = ', l1_cv.coef_)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error

#LAsso reults
print("R2:", r2_score(y_test2, lasso.predict(X_test2)))
print("MAE:", mean_absolute_error(y_test2, lasso.predict(X_test2)))
```

```python
print(list(zip(l1_cv.coef_, X_test2.columns.values)))
```

```python
coeff = list(zip(l1_cv.coef_, X_test2.columns.values))
df_coeff = pd.DataFrame(coeff)
```
